```python
x0=2
y0=4
xp=2.5
n=101
step_h=(xp-x0)/(n-1)

x=np.linspace(x0,xp,n)

y=np.zeros([n])

y[0]=y0
for k in range(1,n):
  y[k]=y[k-1]+step_h*(0.1*(np.sqrt(y[k-1]))+0.4*(np.power(x[k-1],2)))

for k in range(n):
  print(x[k],y[k])
```

```python
plt.plot(x,y,'o')
plt.xlabel("Value of x")
plt.ylabel("Value of y")
plt.title("Approximate Solution with Euler's Method")
plt.show()
```
